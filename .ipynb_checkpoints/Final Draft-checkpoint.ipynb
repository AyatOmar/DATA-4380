# 2. Cleaning book titles

```python
#minimizing search space
UserData["ModifiedName"] = UserData["Name"].str.replace("[^a-zA-Z0-9 ]","", regex =True)
#making it all in lowercase
UserData["ModifiedName"] = UserData["ModifiedName"].str.lower()
#replacing multiple spaces with single space
UserData["ModifiedName"] = UserData["ModifiedName"].str.replace("\s+", " ",regex=True)
```

```python
UserData
```

Delimma ------> my dataset does not assign book IDs  ------> need to assign book IDs to the book titles

```python
# Create a dictionary to map book titles to IDs
book_title_id_map = {}
current_id = 1

# Iterate through the book titles in the DataFrame
for title in UserData['ModifiedName']:
    # If the title is not already in the dictionary, assign a new ID
    if title not in book_title_id_map:
        book_title_id_map[title] = current_id
        current_id += 1

# Create a new column 'book_id' in the DataFrame
UserData['book_id'] = UserData['ModifiedName'].map(book_title_id_map)

# Print the DataFrame to see the 'book_id' column
print(UserData)
```

# 3. Creating a Term Frequency Matrix for search engine


                              potter          harry              the
      
          The                    0              0                 1
          
    Harry The Potter             1              1                 1  
     
     The Harry                   0              1                 1
     
 ---------------> inverse document frequence 
 ---------------> Term frequency Matrix * inverse document frequency
 ---------------> you get similarity scores 
 
 reference: Dataquest on youtube

```python
#this python library does ^ this
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(UserData["ModifiedName"])
```

```python
#compute similarity between entered name and list of name
from sklearn.metrics.pairwise import cosine_similarity

def search(name):
    #search = input("Enter book title: ")
    #cleaning input like how the titles were cleaned in the dataset
    import re
    def clean_search(name):
        return re.sub("[^a-zA-Z0-9 ]", "", name)
    search = clean_search(name)
    #query_vec is comparing the searched title to each of the modified names in the dataset and return how similar they are
    query_vec = vectorizer.transform([name])

    #find similarity between searched term and titles in the data
    #output is numpy vector 
    #searching for titles similar to what was searched
    similarity = cosine_similarity(query_vec, tfidf).flatten()

    #finding titles that are the most similar to what book is being searched
    #this finds the 5 most similar book titles in the dataset to the searched book title
    indices = np.argpartition(similarity, -5)[-5:]
    #indexing the userdata by the indices to obtain titles
    #the last part [::-1] is to make the list output most similar to least 
    results = UserData.iloc[indices][::-1]
    return results
```

```python
#building interactive search box
#type in name of the book and see search results
#widgets are interactive allowing us to embed
import ipywidgets as widgets
#function you can use as output
from IPython.display import display


#creating input widget
book_input = widgets.Text(
    value = "",
    discription = "Book Title: ",
    #it will be inabled?
    disabled = False
)
#output widget
book_list = widgets.Output()
def on_type(data):
    with book_list:
        book_list.clear_output()
        title = data["new"]
        if len(title)>5:
            display(search(title))
    
book_input.observe(on_type, names = 'value')
display(book_input, book_list)
```

# 4. Building recommendation system

finding users who liked a book and seeing what other book titles they also liked, this is collaborative filltering recommendation system 

```python
book_id = 6404
#finding any users that read the book with id 1 and that rated it a 4 or higher
similar_users = UserData[(UserData["book_id"] == book_id) & (UserData["NumericRating"] >= 4)]["ID"].unique()
```

```python
similar_users
```

```python
#finding other books they also liked
similar_user_recs = UserData[(UserData["ID"].isin(similar_users))&(UserData["NumericRating"] >= 4)]["book_id"]
```

```python
#this is the list of books from users that liked the same book who also rated it a 4 or greater
similar_user_recs
```

```python
#converting to percentage
similar_user_recs = similar_user_recs.value_counts() / len(similar_users)
#giving books that were 50% or more liked
similar_user_recs = similar_user_recs[similar_user_recs >= 0.5]
```

```python
#this is a list of books that other users also liked to the searched book (it ends with us)
similar_user_recs
```

```python
#how much all of the users like the book and finding everyone that rated it highly
all_user = UserData[(UserData["book_id"].isin(similar_user_recs.index)) & (UserData["NumericRating"] >= 4)]
```

```python
#all of the users that read the book and rated it with a 4 or greater
all_user
```

```python
#everyone who liked the book it ends with us,but from all the users that liked these books and other similar books
all_user_recs = all_user["book_id"].value_counts() / len(all_user["ID"].unique())
```

```python
all_user_recs
```

```python
#comparing the percentages
rec_percent = pd.concat([similar_user_recs, all_user_recs], axis =1)
rec_percent.columns = ["similar", "all"]
```

```python
#this gives us each of the books that were recommened to us and how many users similar to the book search liked them and how much the average person liked them
#we want the books with the greatest difference
rec_percent
```

```python
rec_percent["score"] = rec_percent["similar"] / rec_percent["all"]
```

```python
rec_percent = rec_percent.sort_values("score", ascending = False)
```

```python
rec_percent
#the score value is giving the ratio between how much similar users liked the book vs how much the average user liked the book
#the higher the score, the better the recommendation
```

```python
Data2 = rec_percent.head(11).merge(UserData, left_index = True, right_on = "book_id")
```
