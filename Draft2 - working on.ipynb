# Data Processing Book Name
This is to make it easier for the search engine, so removing extra characters
ex:

Harry Potter

Harry POTTER

harry potter

is all the same but might not search the same

so doing some data processing can make the search engine more efficient


```python
#minimizing search space
UserData["Modified_Name"] = UserData["Name"].str.replace("[^a-zA-Z0-9 ]","", regex =True)
```

```python
UserData
```

```python
#making it all in lowercase
UserData["Modified_Name"] = UserData["Modified_Name"].str.lower()
```

```python
UserData
```

```python
#replacing multiple spaces with single space
UserData["Modified_Name"] = UserData["Modified_Name"].str.replace("\s+", " ",regex=True)
```

```python
UserData
```

# Creating book search engine


```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
tfidf = vectorizer.fit_transform(UserData["Modified_Name"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import re

def search(title, vectorizer):
    processed = re.sub("[^a-zA-Z0-9 ]","", title.lower())
    title_vec = vectorizer.transform([processed])
    similarity = cosine_similarity(title_vec, tfidf).flatten()

    indices = np.argpartition(similarity,-5)[-5:]
    results = UserData.iloc[indices]
    
    return results
```

```python
search('Harry potter and the prisoner of azkaban',vectorizer)
```

# Recommendations

```python
#using the book search engine from before we find all the users that also read the book
Users = search("Harry Potter and the Half-Blood Prince", vectorizer)
Users
```

```python
import ipywidgets as widgets
from IPython.display import display

book_input = widgets.Text(
    value = "Harry Potter and the prisoner of azkaban",
    description = "Book title:",
    disabled = False
)

book_list = widgets.Output()

def on_type(data):
    with book_list:
        book_list.clear_output()
        title = data["new"]
        if len(title) > 5:
            display(search(title))

book_input.observe(on_type, names='value')

display(book_input, book_list)
```
